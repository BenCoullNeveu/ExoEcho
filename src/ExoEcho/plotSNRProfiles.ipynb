```python
import pandas as pd
from Functions import *
from Telescope import *
import Ariel

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
```

```python
# getting target list

targets = Ariel.getInstrument("VISPhot").target_list
```

```python
wasp_list = targets[targets["Planet Name"].str.contains("WASP")]

for i in range(0, 30):
    target = wasp_list.iloc[i]
    Ariel.plotSNRProfile(target["Planet Name"], default_save=True)
    plt.show()
```
